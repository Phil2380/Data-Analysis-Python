#### As we wish to conduct an analysis to find new markets for chicken exportation, let's select only the rows that correspond to this market

```python
food_avalable_2017=food_avalable_2017[food_avalable_2017['Produit']=='Viande de Volailles']
food_avalable_2017
```

```python
Chicken_2017=pd.pivot(food_avalable_2017,index='Zone',columns='Élément',values='Valeur')
Chicken_2017
```

#### describe Chicken_2017

#### Having already worked on this type of datas, I know that "Disponibilité intérieur" is equal to the sum of "Nouriture", "pertes"et"aliments pour animeaux" I, therefore, will delete those columns as well as the columns Semences, "Autres utilisations (non alimentaire)","Traitement", "Alimentation pour touristes"columns that do not have sufficient data quantity. The columns "Résidus", "Disponibilité de matière grasse en quantité (g/personne/jour)" and "Disponibilité de protéines en quantité (g/personne/jour)" won't be needed either for the purpose of the analysis

```python
Chicken_2017=Chicken_2017.drop(['Nourriture','Pertes','Aliments pour animaux','Autres utilisations (non alimentaire)','Traitement','Semences','Résidus','Alimentation pour touristes','Disponibilité de matière grasse en quantité (g/personne/jour)','Disponibilité de protéines en quantité (g/personne/jour)'],axis=1)
```
